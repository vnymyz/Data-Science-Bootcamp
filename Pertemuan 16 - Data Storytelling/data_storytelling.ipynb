## 1. Data Story telling

Data story telling adalah gabungan antara data, visualisasi dan narasi.

Data story telling itu juga bukan hanya membuat grafik tapi bagaimana kita bisa menyampaikan pesan
atau insight dari data melalui visualisasi yang jelas, menarik dan bermakna.

Jadi Data Story telling itu gabungan dari :

- Data (fakta & angka)

- Visualisasi (grafik, chart, dll.)

- Narasi (penjelasan yang mudah dimengerti)

Fokus utamanya adalah komunikasi data secara efektif, bukan hanya teknis coding grafik.

Data story telling itu penting karena :
- membuat data mudah dimengerti.

- mempengaruhi kebutuhan bisnis 

- Agar audiens (termasuk non-teknis) memahami data yang cepat, tepat dan mudah diingat.

Story telling yang buruk : grafik rumit tanpa penjelasan.

Story telling yang baik : grafik sederhana, insight jelas.

Contoh Narasi Pendek :
"Grafik ini menunjukkan bahwa penumpang kelas 3 Titanic jauh lebih banyak dari kelas lain, namun tingkat keselamatannya rendah."

## 2. Jenis Grafik dan Fungsinya

| Jenis Grafik | Fungsi                | Contoh                     |
| ------------ | --------------------- | -------------------------- |
| Bar Chart    | Perbandingan kategori | Jumlah penumpang per kelas |
| Line Chart   | Perubahan waktu       | Tren penjualan             |
| Pie Chart    | Komposisi             | Persentase jenis kelamin   |
| Histogram    | Distribusi data       | Usia penumpang             |
| Boxplot      | Outlier & persebaran  | Usia per kelas             |
| Heatmap      | Korelasi variabel     | Korelasi Fare & Survived   |


#### Prinsip Desain Visualisasi
- Pilih grafik yang tepat

- Minimalis (hindari clutter)

- Warna untuk makna, bukan dekorasi

- Highlight insight penting.

- Fokus pada apa yang penting untuk dibaca.

## 3. Eksplorasi Dataset

```python
import pandas as pd
import seaborn as sns

# Membaca file CSV
df = pd.read_csv('train.csv')

# Melihat 5 baris pertama
df.head()
```

#### Barplot - Mengetahui jumlah penumpang per kelas

```python
sns.countplot(x='Pclass', data=df)
# "Penumpang paling banyak berada di kelas 3."
```

#### Pie Chart - Komposisi Jenis Kelamin

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
# "Mayoritas penumpang adalah laki-laki."
```

#### Boxplot - Distribusi Usia per Kelas

```python
sns.boxplot(x='Pclass', y='Age', data=df)
# "Kelas 1 memiliki penumpang usia lebih tua."
```

#### Heatmap - Korelasi antar variabel numerik

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
# "Tiket lebih mahal cenderung lebih tinggi peluang selamat."
```

```python
# buat cek ada kolom apa aja
print(df.columns.tolist())
```

```python
#  12. Heatmap Korelasi
import seaborn as sns
import matplotlib.pyplot as plt

# Ambil kolom numerik yang relevan
numerik = df[['fare', 'age', 'sibsp', 'parch', 'survived']]
corr = numerik.corr()

# Visualisasi heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Variabel Numerik')
plt.show()
```

#### Insight:

- Korelasi tertinggi terhadap Survived adalah Fare (0.26).

- Korelasi negatif kecil dengan SibSp dan Parch, artinya jumlah saudara/orangtua tidak terlalu berpengaruh kuat terhadap keselamatan.

```python
# 13. Scatterplot Fare vs Age
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='age', y='fare', hue='survived', palette='coolwarm')
plt.title('Scatterplot Umur vs Harga Tiket')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

#### Insight:

- Tidak ada pola linier yang jelas.

- Sebagian besar penumpang membayar di bawah $100.

- Ada beberapa penumpang muda dan tua yang membayar mahal.

```python
# 14. Stripplot Survived vs Fare (Hue: Sex)
plt.figure(figsize=(8,6))
sns.stripplot(data=df, x='survived', y='fare', hue='sex', jitter=True, dodge=True)
plt.title('Harga Tiket dan Keselamatan berdasarkan Jenis Kelamin')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.ylabel('Fare')
plt.show()
```

#### Insight:

- Penumpang wanita yang selamat lebih banyak ditemukan membeli tiket mahal.

- Penumpang pria banyak di kedua sisi, tapi yang selamat cenderung membayar lebih rendah.

- Tiket mahal kemungkinan berasal dari kelas 1 (First Class).
