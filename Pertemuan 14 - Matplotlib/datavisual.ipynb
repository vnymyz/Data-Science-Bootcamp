Matplotlib adalah fondasi dari semua visualisasi di Python

Matplotlib adalah library visualisasi data di Python yang paling populer dan serbaguna. Library ini memungkinkan kamu untuk membuat berbagai jenis grafik seperti:
- Line chart (grafik garis)

- Bar chart (grafik batang)

- Pie chart (diagram lingkaran)

- Scatter plot (grafik sebar)

- Heatmap, dan lainnya

### Matplotlib juga sering digunakan dalam :

### Matplotlib memiliki Hierarki (Struktur Matplotlib)
| Komponen   | Penjelasan                                       |
| ---------- | ------------------------------------------------ |
| **Figure** | Kanvas utama tempat semua grafik dibuat          |
| **Axes**   | Area dalam figure tempat grafik ditempatkan      |
| **Axis**   | Sumbu X dan Y dari axes                          |
| **Artist** | Semua elemen visual (garis, teks, legenda, dll.) |

### Kelebihan Matplotlib

### 4. Visualisasi Dasar

### Line Chart
Grafik yang menghubungkan titik-titik data dengan garis.

Sumbu X biasanya digunakan untuk bulan, hari, tahun.

Sumbu Y biasanya untuk nilai yang diukur seperti  penjualan, suhu.

Cocok untuk melihat perubahan trending.

#### Kalau dalam Data Science digunakan untuk :
- Melihat tren pertumbuhan/penurunan suatu variabel (contoh: penjualan harian, traffic website).

- Menunjukkan musiman (seasonal pattern).

- Memonitor performa dari waktu ke waktu.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = {
    "Bulan": ["Jan", "Feb", "Mar", "Apr", "Mei"],
    "Penjualan": [150, 200, 180, 220, 210],
    "Kategori A": [50, 80, 60, 90, 70],
    "Kategori B": [100, 120, 120, 130, 140]
}

plt.plot(data["Bulan"], data["Penjualan"], marker='o', color='blue')
plt.title("Pertumbuhan Penjualan per Bulan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah")
plt.grid(True)
plt.show()
````

Membandingkan tren penjualan Kategori A dan Kategori B dari bulan ke bulan dalam satu grafik garis (line chart) :

````python
# Line chart untuk Kategori A dan B
plt.plot(data["Bulan"], data["Kategori A"], marker='o', color='orange', label="Kategori A")
plt.plot(data["Bulan"], data["Kategori B"], marker='s', color='green', label="Kategori B")

# Judul dan label
plt.title("Tren Penjualan Kategori A dan B")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penjualan")
plt.grid(True)
plt.legend()  # Menampilkan keterangan warna garis

plt.tight_layout()
plt.show()
````

### Bar Chart

Grafik batang tegak (vertikal) atau datar (horizontal) untuk membandingkan nilai antar kategori.

Sumbu X (atau Y) biasanya Nama kategori (misalnya: produk A, B, C).

Sumbu Y (atau X) biasanya Nilai (jumlah, persentase, dll).

Cocok untuk melakukan Perbandingan antar kategori.

#### Kalau dalam Data Science digunakan untuk :
- Membandingkan performa antar kelompok (contoh: produk A vs produk B).

- Menunjukkan distribusi jumlah data dalam kategori.

- Visualisasi hasil agregasi (contoh: groupby() hasil penjumlahan Sales per kategori).

````python
data = {
    "Bulan": ["Jan", "Feb", "Mar", "Apr", "Mei"],
    "Penjualan": [150, 200, 180, 220, 210],
    "Kategori A": [50, 80, 60, 90, 70],
    "Kategori B": [100, 120, 120, 130, 140]
}
````

````python
plt.bar(data["Bulan"], data["Kategori A"], color='skyblue')

# Menambahkan judul dan label sumbu
plt.title("Total Penjualan Kategori A")
plt.xlabel("Bulan")
plt.ylabel("Jumlah")

# Menambahkan angka di atas setiap bar
for i, value in enumerate(data["Kategori A"]):
    plt.text(i, value + 2, str(value), ha='center', fontsize=10, color='red')

plt.show()
````

### Kalau mau munculin kategori A dan B di bar chart :

````python
bulan = data["Bulan"]
kategori_a = data["Kategori A"]
kategori_b = data["Kategori B"]

# Tentukan posisi bar di x-axis
x = np.arange(len(bulan))  # posisi dasar
width = 0.35  # lebar bar

# Buat plot
plt.figure(figsize=(8, 5))
plt.bar(x - width/2, kategori_a, width, label='Kategori A', color='orange')
plt.bar(x + width/2, kategori_b, width, label='Kategori B', color='skyblue')

# Tambahkan label dan judul
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penjualan")
plt.title("Total Penjualan Kategori A dan B per Bulan")
plt.xticks(x, bulan)  # Ganti posisi x-axis dengan nama bulan
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

### Penjelasan :
- np.arange(len(bulan)): Membuat titik posisi untuk batang bar.

- width: Mengatur lebar setiap batang.

- x - width/2 dan x + width/2: Menggeser posisi bar agar tidak tumpang tindih (jadi sejajar).

- plt.legend(): Menampilkan label untuk masing-masing kategori.

### Scatter plot

Scatter plot adalah grafik yang digunakan untuk menampilkan hubungan (relasi) antara dua variabel numerik (angka). Data ditampilkan dalam bentuk titik-titik (dot) di bidang 2 dimensi (x dan y).

Grafik berupa titik-titik yang menunjukkan hubungan (korelasi) antara dua variabel numerik.

Sumbu X dan Y Masing-masing adalah nilai dari dua variabel.

Scatter plot membantu melihat apakah dua variabel saling berhubungan atau tidak. Kamu bisa menggunakannya untuk mendeteksi pola, outlier (titik aneh), dan arah hubungan antara dua variabel numerik.

Cocok untuk Mendeteksi hubungan atau pola antara dua variabel.

#### Dalam Data Science Scatter Plot digunakan untuk :
- Mendeteksi korelasi positif/negatif antar dua variabel.

- Mencari anomali (outlier).

- Membantu pemodelan regresi (contoh: prediksi harga berdasarkan ukuran rumah).

### Cara Membaca Scatter Plot
#### 1. Setiap titik = 1 data

- Sumbu X → nilai dari variabel pertama.

- Sumbu Y → nilai dari variabel kedua.

- Setiap titik adalah pasangan nilai (x, y).

Contoh :

```plaintext
x = Berat Badan
y = Tinggi Badan
Tiap titik = 1 orang
```

#### 2. Perhatikan Pola Titik
| Pola Titik                      | Arti                                        |
| ------------------------------- | ------------------------------------------- |
| 🔼 **Naik ke kanan** (positif)  | Jika X naik, Y juga naik → hubungan positif |
| 🔽 **Turun ke kanan** (negatif) | Jika X naik, Y turun → hubungan negatif     |
| 🔹 **Acak / tersebar rata**     | Tidak ada hubungan yang jelas               |
| 🔸 **Titik rapat**              | Hubungan kuat                               |
| 🔸 **Titik menyebar jauh**      | Hubungan lemah                              |

#### 3. Lihat Korelasi (Trend)
- Hubungan Positif :
```markdown
         .
       .
     .
   .
```
- Hubungan Negatif :
```markdown
   .
     .
       .
         .
```

- Enggak ada korelasi atau hubungan :
```markdown
  .  .       .
   .     . .
      .    .
```

#### 4. Gunakan Warna/Ukuran untuk Dimensi Tambahan
- Warna bisa mewakili kategori (misal: jenis kelamin, hari, tipe barang).

- Ukuran bisa menunjukkan nilai tambahan (misal: jumlah pembelian).

#### Analogi
Misal kita punya grafik dimana :
- X = Jumlah jam belajar

- Y = Nilai ujian

Kalau titik-titik cenderung naik ke kanan artinya :

*Semakin banyak jam belajar, semakin tinggi nilai ujian (hubungan positif).*

Perbandingan antara Kategori A dan Kategori B untuk setiap bulan dalam bentuk scatter plot.

### Tebak Hubungan Antar Variabel (Scatter Plot)
Sebuah survei dilakukan terhadap 15 orang siswa. Data yang dikumpulkan meliputi:

- Jumlah jam belajar per minggu (X)

- Nilai ujian akhir (Y)

Berikut scatter plot-nya:

### Scatter plot hubungan positif

````python
# Dataset jam belajar siswa dengan nilai ujian
import matplotlib.pyplot as plt

# Contoh data: Jam Belajar vs Nilai Ujian
jam_belajar = [1, 2, 3, 4, 5, 6]
nilai_ujian = [40, 50, 60, 70, 80, 90]

plt.figure(figsize=(6,4))
plt.scatter(jam_belajar, nilai_ujian, color='green')
plt.title("Hubungan Positif: Jam Belajar vs Nilai Ujian")
plt.xlabel("Jam Belajar")
plt.ylabel("Nilai Ujian")
plt.grid(True)
plt.show()
````

Titik-titik naik ke kanan atas → menunjukkan bahwa semakin banyak belajar, semakin tinggi nilai.

### Scatter Plot dengan Hubungan Negatif


````python
jumlah_rokok = [1, 2, 3, 4, 5, 6]
kapasitas_paru = [90, 85, 80, 70, 60, 50]

plt.figure(figsize=(6,4))
plt.scatter(jumlah_rokok, kapasitas_paru, color='red')
plt.title("Hubungan Negatif: Rokok vs Kapasitas Paru-paru")
plt.xlabel("Jumlah Rokok per Hari")
plt.ylabel("Kapasitas Paru-paru (%)")
plt.grid(True)
plt.show()
````

Titik-titik menurun ke kanan bawah → semakin banyak merokok, semakin turun kapasitas paru-paru.

### Scatter Plot yang tidak ada hubungannya

````python
# Scatter plot Kategori A vs Kategori B
plt.figure(figsize=(8,5))
plt.scatter(data["Kategori A"], data["Kategori B"], color='purple', marker='o', s=100)

# Tambahkan label titik dengan nama bulan
for i in range(len(data["Bulan"])):
    plt.text(data["Kategori A"][i] + 1, data["Kategori B"][i], data["Bulan"][i])

# Judul dan Label
plt.title("Scatter Plot: Kategori A vs Kategori B")
plt.xlabel("Penjualan Kategori A")
plt.ylabel("Penjualan Kategori B")
plt.grid(True)
plt.show()
````

### Penjelasan :
- plt.scatter() digunakan untuk membuat scatter plot.

- marker='o' menunjukkan bentuk titik (lingkaran).

- s=100 mengatur ukuran titik.

- plt.text() menambahkan label nama bulan di dekat titiknya.

- Ini membantu untuk melihat hubungan antara dua kategori produk.

### Penjelasan Scatter plot nya :
- Jan: (50, 100)

- Feb: (80, 120)

- Mar: (60, 120)

- Apr: (90, 130)

- Mei: (70, 140)

- Kadang nilai Kategori A naik, tapi Kategori B nggak selalu ikut naik atau turun.

- Hubungan antar sumbu X dan Y-nya tidak konsisten.

## Implementasi Matplotlib dengan Dataset Netflix

````python
# Import library dulu kek biasa
import pandas as pd
import matplotlib.pyplot as plt

# Load data netflix
df = pd.read_csv("netflix_titles.csv")
````

### 1. Bar Chart - Jumlah Tayangan Berdasarkan Tahun Rilis

````python
# Hitung jumlah tayangan berdasarkan tahun
release_counts = df['release_year'].value_counts().sort_index()

# Buat bar chart
plt.figure(figsize=(15,5))
plt.bar(release_counts.index, release_counts.values, color='skyblue')
plt.title("Jumlah Tayangan Netflix Berdasarkan Tahun Rilis", fontsize=14)
plt.xlabel("Tahun Rilis", fontsize=12)
plt.ylabel("Jumlah Tayangan", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

### 2. Line Chart - Pertumbuhan Jumlah Tayangan Netflix per Tahun
bagaimana pertumbuhan jumlah konten Netflix dari tahun ke tahun.

#### Line chart cocok digunakan untuk:

- Menampilkan perubahan data dari waktu ke waktu.

- Cocok untuk tren tahunan, bulanan, atau harian.

````python
# Hitung jumlah tayangan berdasarkan tahun rilis
release_counts = df['release_year'].value_counts().sort_index()

# Buat Line Chart
plt.figure(figsize=(12,5))
plt.plot(release_counts.index, release_counts.values, marker='o', linestyle='-', color='green')

# Tambahkan judul dan label
plt.title("Pertumbuhan Jumlah Tayangan Netflix per Tahun", fontsize=14)
plt.xlabel("Tahun Rilis", fontsize=12)
plt.ylabel("Jumlah Tayangan", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

### 3. Scatter Plot - Cek Korelasi
mengecek apakah durasi tayangan memiliki hubungan dengan tahun rilis. Maka kita bisa memplot:

- Sumbu X: Tahun rilis

- Sumbu Y: Durasi tayangan (dalam menit) — hanya berlaku untuk tipe Movie

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# Filter hanya Movie
movies_df = df[df['type'] == 'Movie'].copy()

# Ambil durasi angka (misal '90 min' → 90)
movies_df['duration_num'] = movies_df['duration'].str.extract(r'(\d+)').astype(float)

# Hapus data yang durasi atau tahun rilis-nya kosong
movies_df = movies_df.dropna(subset=['duration_num', 'release_year'])
````

````python
# Data Visualisasi Scatter Plot
plt.figure(figsize=(12,6))
plt.scatter(movies_df['release_year'], movies_df['duration_num'], alpha=0.5, color='tomato')
plt.title("Scatter Plot: Durasi Film vs Tahun Rilis", fontsize=14)
plt.xlabel("Tahun Rilis", fontsize=12)
plt.ylabel("Durasi Film (menit)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

### Penjelasan dari Scatter Plot tersebut :

### 5. Box Plot Durasi Film Berdasarkan Rating

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# Filter hanya Movie
movies_df = df[df['type'] == 'Movie'].copy()

# Ambil angka dari kolom durasi (misal: '90 min' → 90)
movies_df['duration_num'] = movies_df['duration'].str.extract(r'(\d+)').astype(float)

# Hapus baris yang durasi atau rating-nya kosong
movies_df = movies_df.dropna(subset=['duration_num', 'rating'])

# Ambil hanya beberapa rating populer
rating_umum = ['PG', 'PG-13', 'R', 'TV-MA', 'TV-14']
filtered_df = movies_df[movies_df['rating'].isin(rating_umum)]

# Buat Box Plot
plt.figure(figsize=(12,6))
filtered_df.boxplot(column='duration_num', by='rating', grid=True, patch_artist=True,
                    boxprops=dict(facecolor='lightskyblue', color='black'),
                    medianprops=dict(color='red'),
                    whiskerprops=dict(color='green'),
                    capprops=dict(color='purple'))
plt.title("Box Plot Durasi Film berdasarkan Rating", fontsize=14)
plt.suptitle("")  # Menghapus judul otomatis dari pandas
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Durasi (menit)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

### Penjelasan Visualisasi Box Plot

### 🔷 1. Kotak Biru (Box)
- Kotak biru menunjukkan rentang interkuartil (IQR), yaitu rentang dari kuartil pertama (Q1) ke kuartil ketiga (Q3):

- Batas bawah kotak: Kuartil pertama (Q1), yaitu 25% data di bawah nilai ini.

- Batas atas kotak: Kuartil ketiga (Q3), yaitu 75% data di bawah nilai ini.

- Ini mencakup 50% tengah dari data durasi film dalam tiap kategori rating.

### 🔴 2. Garis Merah di dalam Kotak
- Garis merah ini menunjukkan median (Q2), yaitu nilai tengah data:

- Artinya, 50% data lebih kecil dan 50% lebih besar dari nilai ini.

- Garis merah bisa tidak berada di tengah kotak jika datanya tidak simetris.

### ⚫ 3. Garis Hitam di atas dan bawah kotak (Whiskers)
- Disebut juga "whiskers", ini mewakili rentang data yang bukan outlier:

- Garis atas: titik data maksimum yang masih dalam rentang wajar.

- Garis bawah: titik data minimum yang masih dalam rentang wajar.

- Biasanya dihitung hingga 1.5 × IQR dari Q1 ke bawah dan Q3 ke atas.

### 4. Lingkaran Kecil (Outliers)
- Lingkaran-lingkaran kecil ini adalah outlier:

- Titik data yang berada di luar rentang whiskers, alias nilai-nilai ekstrim.

- Contohnya: film berdurasi sangat pendek (misalnya < 20 menit) atau sangat panjang (> 200 menit).

- Mereka tetap penting karena bisa mengindikasikan film spesial seperti film pendek atau dokumenter panjang.

### Contoh Interpretasi Rating “TV-MA”:
- Median durasi film berada di sekitar 100 menit.

- IQR antara kira-kira 80 hingga 120 menit.

- Ada banyak outlier, termasuk film yang berdurasi lebih dari 300 menit, yang sangat tidak umum.

### Kesimpulan Umum:
- Semua rating memiliki persebaran durasi yang mirip secara umum.

- Kategori seperti TV-14 memiliki rentang durasi yang lebih besar (IQR lebih tinggi).

- Outlier cukup banyak, artinya dalam dataset kamu terdapat banyak film berdurasi tidak umum, baik terlalu pendek maupun terlalu panjang.

### 7. Stacked Bar Plot – Jumlah Movie dan TV Show per Country (Top 5)

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# Ambil 5 negara dengan jumlah konten terbanyak
top5_countries = df['country'].value_counts().head(5).index

# Filter data hanya negara top 5
top_df = df[df['country'].isin(top5_countries)]

# Buat tabel pivot: jumlah Movie & TV Show per negara
pivot = top_df.pivot_table(index='country', columns='type', aggfunc='size', fill_value=0)

# Plot stacked bar
pivot.plot(kind='bar', stacked=True, figsize=(10,6), color=['skyblue', 'salmon'])
plt.title("Jumlah Movie & TV Show di 5 Negara Teratas", fontsize=14)
plt.xlabel("Negara")
plt.ylabel("Jumlah Konten")
plt.xticks(rotation=45)
plt.legend(title="Tipe")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

### Kesimpulan dari semua chart dalam Matplotlib

| **Jenis Chart**          | **Fungsi Utama**                                                                  | **Cocok Digunakan Untuk**                                     | **Contoh Penggunaan di Dunia Kerja Data Science**     |
| ------------------------ | --------------------------------------------------------------------------------- | ------------------------------------------------------------- | ----------------------------------------------------- |
| **Bar Chart**            | Menampilkan perbandingan antar kategori.                                          | Data kategorikal dengan jumlah per kategori.                  | Menampilkan jumlah pelanggan berdasarkan provinsi.    |

````python
# No 5
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  Baca dataset
df = pd.read_csv("netflix_titles.csv")

#  Filter hanya yang bertipe Movie
df_movies = df[df['type'] == 'Movie'].copy()

#  Ekstrak angka durasi film dalam satuan menit
df_movies['duration_num'] = df_movies['duration'].str.extract(r'(\d+)').astype(float)

#  Buat Box Plot dengan Grid
plt.figure(figsize=(10, 6))
sns.boxplot(x="rating", y="duration_num", data=df_movies)
plt.title("Durasi Film Berdasarkan Rating")
plt.xlabel("Rating")
plt.ylabel("Durasi (menit)")
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', linewidth=0.7)  # Tambahkan grid horizontal
plt.tight_layout()
plt.show()

````

````python
# No.6
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset Netflix
df = pd.read_csv("netflix_titles.csv")

# Filter hanya yang bertipe Movie
df_movies = df[df["type"] == "Movie"].copy()

# Daftar genre yang ingin dicari
genre_keywords = ['Drama', 'Comedy', 'Action', 'Documentary']

# Hitung jumlah film untuk masing-masing genre
genre_counts = {}
for genre in genre_keywords:
    genre_counts[genre] = df_movies["listed_in"].str.contains(genre, na=False).sum()

# Visualisasi dalam bentuk Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(
    genre_counts.values(),
    labels=genre_counts.keys(),
    autopct='%1.2f%%',
    startangle=140,
    colors=['gold', 'lightblue', 'lightgreen', 'pink']
)
plt.title("Proporsi Genre Utama pada Film")
plt.tight_layout()
plt.show()

# 6. Tampilkan hasil count per genre
print(genre_counts)
````

````python
# no.7
import pandas as pd
import matplotlib.pyplot as plt

#  Baca dataset Netflix
df = pd.read_csv("netflix_titles.csv")

#  Hapus baris dengan rating atau type yang kosong
df_clean = df.dropna(subset=["rating", "type"])

#  Kelompokkan berdasarkan rating dan type, lalu hitung jumlahnya
type_rating_counts = df_clean.groupby(["rating", "type"]).size().unstack().fillna(0)

#  Plot stacked bar chart
plt.figure(figsize=(10, 6))
type_rating_counts.plot(kind='bar', stacked=True, colormap='Set2')
plt.title("Jumlah Movie/TV Show Berdasarkan Rating")
plt.xlabel("Rating")
plt.ylabel("Jumlah")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# No.8
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset Netflix
df = pd.read_csv("netflix_titles.csv")

# Bersihkan data: hapus baris tanpa informasi negara
df_clean = df.dropna(subset=["country"])

# Hitung 10 negara teratas berdasarkan jumlah film/TV Show
top_countries = df_clean["country"].value_counts().head(10)

# Plot bar horizontal
plt.figure(figsize=(10, 6))
top_countries.sort_values().plot(kind='barh', color='orchid')
plt.title("10 Negara dengan Jumlah Film/TV Show Terbanyak")
plt.xlabel("Jumlah")
plt.ylabel("Negara")
plt.tight_layout()
plt.show()
````
