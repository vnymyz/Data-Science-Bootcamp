## Implementasi Data Netflix menggunakan Seaborn

### 1. Distribusi (histplot, kdeplot)

```python
# Import library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca dataset
df = pd.read_csv("netflix_titles.csv")

# Filter hanya yang bertipe 'Movie'
df_movies = df[df['type'] == 'Movie'].copy()

# Ekstrak angka durasi dalam menit
df_movies['duration_num'] = df_movies['duration'].str.extract(r'(\d+)').astype(float)

# Plot distribusi durasi film
plt.figure(figsize=(10, 6))
sns.histplot(df_movies["duration_num"], kde=True, bins=100, color="salmon")
plt.title("Distribusi Durasi Film Netflix")
plt.xlabel("Durasi (menit)")
plt.ylabel("Frekuensi")
plt.tight_layout()
plt.show()
```

## 2. Boxplot: Perbandingan durasi antar rating

```python
# 1. Import library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load dataset Netflix
df = pd.read_csv("netflix_titles.csv")

# 3. Filter data: hanya 'Movie'
df_movies = df[df['type'] == 'Movie'].copy()

# 4. Ekstrak angka durasi (dalam menit)
df_movies['duration_num'] = df_movies['duration'].str.extract(r'(\d+)').astype(float)

# 5. Hapus baris dengan rating atau durasi kosong
df_movies.dropna(subset=['rating', 'duration_num'], inplace=True)

# 6. Visualisasi Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="rating", y="duration_num", data=df_movies)
plt.title("Durasi Film Berdasarkan Rating")
plt.xlabel("Rating")
plt.ylabel("Durasi (menit)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3.  Barplot: Rata-rata durasi tiap genre tertentu (pakai keyword)

```python
# 1. Import library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 2. Load dataset Netflix
df = pd.read_csv("netflix_titles.csv")

# 3. Filter data: hanya 'Movie'
df_movies = df[df['type'] == 'Movie'].copy()

# 4. Ekstrak angka durasi dari kolom 'duration'
df_movies['duration_num'] = df_movies['duration'].str.extract(r'(\d+)').astype(float)

# 5. Hapus baris yang kolom 'listed_in' atau 'duration_num' kosong
df_movies.dropna(subset=['listed_in', 'duration_num'], inplace=True)

# 6. Genre yang ingin dianalisis
genre_keywords = ['Drama', 'Comedy', 'Action']
avg_durations = []

# 7. Hitung rata-rata durasi untuk setiap genre
for genre in genre_keywords:
    mask = df_movies["listed_in"].str.contains(genre)
    avg_durations.append(df_movies[mask]["duration_num"].mean())

# 8. Visualisasi
plt.figure(figsize=(8, 5))
sns.barplot(x=genre_keywords, y=avg_durations, palette='pastel')
plt.title("Rata-rata Durasi Film per Genre")
plt.ylabel("Durasi (menit)")
plt.tight_layout()
plt.show()
```

## 4. Countplot: Jumlah film per rating

```python
# 1. Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load dataset
df = pd.read_csv("netflix_titles.csv")

# 3. Data cleaning
df_movies = df[df["type"] == "Movie"].copy()  # hanya ambil film, bukan TV Show
df_movies = df_movies.dropna(subset=["duration", "rating"])  # hapus data yang tidak ada durasi atau rating

# Ekstrak angka dari kolom durasi (misal: '90 min' jadi 90)
df_movies["duration_num"] = df_movies["duration"].str.extract(r'(\d+)').astype(float)

# 4. Visualisasi Countplot
plt.figure(figsize=(10, 6))
sns.countplot(x="rating", data=df_movies, order=df_movies["rating"].value_counts().index)
plt.title("Jumlah Film per Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. Relasi tahun rilis dengan durasi

```python
# 1. Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load dataset
df = pd.read_csv("netflix_titles.csv")

# 3. Data cleaning
df_movies = df[df["type"] == "Movie"].copy()  # hanya ambil data film
df_movies = df_movies.dropna(subset=["duration", "release_year"])  # hapus data kosong di durasi / tahun rilis

# 4. Ubah durasi ke format numerik
df_movies["duration_num"] = df_movies["duration"].str.extract(r'(\d+)').astype(float)

# 5. Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="release_year", y="duration_num", data=df_movies, alpha=0.4)
plt.title("Tahun Rilis vs Durasi Film")
plt.xlabel("Tahun Rilis")
plt.ylabel("Durasi (menit)")
plt.tight_layout()
plt.show()
```

## 6. Lineplot: Jumlah film tiap tahun


```python
# 1. Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset
df = pd.read_csv("netflix_titles.csv")

# 3. Data Cleaning - Filter hanya Movie
df_movies = df[df["type"] == "Movie"].copy()
df_movies = df_movies.dropna(subset=["release_year"])  # pastikan release_year tidak kosong

# 4. Hitung jumlah film tiap tahun
yearly_counts = df_movies["release_year"].value_counts().sort_index()

# 5. Visualisasi dengan Lineplot
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o')
plt.title("Jumlah Film Tiap Tahun")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Film")
plt.tight_layout()
plt.show()
```

## 7. Heatmap: Korelasi numerik (jika ada)

```python
# 1. Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset
df = pd.read_csv("netflix_titles.csv")

# 3. Data Cleaning - Filter Movie dan proses durasi
df_movies = df[df["type"] == "Movie"].copy()
df_movies = df_movies.dropna(subset=["duration", "release_year"])

# 4. Ekstrak durasi dalam bentuk angka
df_movies["duration_num"] = df_movies["duration"].str.extract(r'(\d+)').astype(float)

# 5. Korelasi dan Visualisasi Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df_movies[["release_year", "duration_num"]].corr(), annot=True, cmap="coolwarm")
plt.title("Korelasi antar Fitur Numerik")
plt.tight_layout()
plt.show()
```

## 8. FacetGrid: Multi-plot berdasarkan kategori (Distribusi Durasi Film per Rating)

```python
# 1. Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load Dataset
df = pd.read_csv("netflix_titles.csv")

# 3. Filter hanya Movie dan buang nilai yang kosong
df_movies = df[df["type"] == "Movie"].copy()
df_movies = df_movies.dropna(subset=["duration", "rating"])

# 4. Ekstrak angka dari kolom durasi
df_movies["duration_num"] = df_movies["duration"].str.extract(r'(\d+)').astype(float)

# 5. Plot distribusi durasi berdasarkan rating menggunakan FacetGrid
plt.figure()
g = sns.FacetGrid(df_movies, col="rating", col_wrap=4, height=3)
g.map(sns.histplot, "duration_num")
g.figure.subplots_adjust(top=0.9)
g.figure.suptitle("Distribusi Durasi Film per Rating")
plt.tight_layout()
plt.show()
```
