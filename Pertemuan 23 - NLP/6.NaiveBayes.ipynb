## Naive Bayes Algorithm

Naive Bayes adalah algoritma machine learning berbasis probabilitas yang digunakan untuk klasifikasi.
Dasarnya adalah Teorema Bayes, yaitu menghitung probabilitas suatu kelas (misalnya ‚ÄúYes‚Äù atau ‚ÄúNo‚Äù) berdasarkan data fitur yang ada.

Disebut "Naive" karena algoritma ini mengasumsikan bahwa setiap fitur independen satu sama lain (padahal di dunia nyata seringkali tidak sepenuhnya independen).

### Manfaat Naive Bayes

- Cepat & Efisien ‚Üí perhitungannya sederhana, cocok untuk dataset besar.

- Akurasi cukup bagus meski asumsi independen sering tidak terpenuhi.

- Mudah diimplementasikan (bisa dikerjakan manual atau pakai library Python seperti sklearn).

- Tidak butuh data training terlalu besar ‚Üí tetap bisa bekerja walaupun dataset terbatas.

### Biasa digunakan dalam kasus

- Sentiment Analysis ‚Üí menentukan apakah suatu review / komentar bernada positif atau negatif.

- Spam Detection ‚Üí memfilter email menjadi spam atau bukan spam.

- Text Classification ‚Üí mengelompokkan dokumen ke dalam kategori tertentu (misalnya berita olahraga, politik, teknologi).

- Medical Diagnosis (klasifikasi sederhana) ‚Üí misalnya apakah pasien berisiko terkena suatu penyakit berdasarkan gejala.

Dalam Naive Bayes kita perlu mengetahui beberapa konsep :

1. Probabilitas atau Probability

### Probability

Dalam probability terdapat 2 event yaitu :

1. Indenpendent Events = dia enggak bakal terjadi apa2

2. Dependent Events = kalau sebelumnya terjadi sesuatu selanjutnya dia bakal terpengaruh.

#### Indenpendent Events

Jika suatu event terjadi ya enggak perubahan ke event berikutnya.

Semisal kita kocok dadu {1,2,3,4,5,6}

Probabilitas :

$$
\Pr(1) = \frac{1}{6}
$$

$$
\Pr(2) = \frac{1}{6}
$$

$$
\Pr(3) = \frac{1}{6}
$$

### Dependent Events

Disebut dengan Dependent Events karena 1 event terjadi suatu perubahan probabilitas selanjutnya juga bakal berubah untuk event selanjutnya. 

Semisal kita ada kotak yang isi nya bola marbel.

```bash
1. 3 bola marbel (orange)

2. 2 bola marbel (kuning)
```

```mathematica
+-----------------+
|   O   O   O     |
|   W       W     |
+-----------------+
```

üü† O = marbel warna oren

‚ö™ Y = marbel warna kuning

Berapa probabilitas nya jika kita membuang 1 bola marbel orange dan 1 marbel kuning ?

- Buang marbel oren :

Probabilitas :

1. Event Pertama. Marbel oren sudah dibuang.
$$
\Pr(0) = \frac{3}{5}
$$


```mathematica
+-----------------+
|      O   O      |
|      W   W      |
+-----------------+
```

2. Event kedua. Probabilitas dari marbel kuning itu berapa setelah event sebelumnya ?

    membuang marbel kuning

$$
\Pr(4) = \frac{2}{4} = \frac{1}{2}
$$

Jadi : 
```bash
P( O and Y) = P(O) * | P(Y/O) | --> Conditional Probability
```

Hasilnya :


$$
\frac{3}{5} \times \frac{1}{2} = \frac{3}{10}
$$


Keterangan :

Probabilitas Orange dikali dengan Probabilitas kuning

- P(O) = Probabilitas Orange. Event Pertama.

- P(Y/O) = Probabilitas si kuning setelah event sebelumnya atau setelah bola orange di ambil. Event Kedua.

Rumus umum dependent event :

$$
P(\text{A and B}) = P(A) \times P\!\left(\frac{B}{A}\right)
$$


Simple nya Kejadian awal * Kejadian sekarang/kejadian sebelumnya

## Teori Bayes

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Keterangan :

- P(A/B) = Probabilitas Event A, setelah Event B sudah terjadi.

- P(A) = Probabilitas Event A.

- P(B) = Probabilitas Event B.

- P(B/A) = Probabilitas Event B, setelah Event A sudah terjadi.

### Independent Feature

```bash
x1          x2          x3              y

-           -           -              Yes

-           -           -              No

-           -           -              Yes

-           -           -              No
```

Keterangan :

- x1 sampai x3 = Fitur Indenpendent

- y = Fitur Dependent

$$
P(\frac{Y}{x_1, x_2, x_3}) = \frac{P(Y) \cdot P\!\left(\frac{x_1, x_2, x_3}{Y}\right)}{P(x_1, x_2, x_3)}
$$

Ini dia dibagi2 lagi :

$$
\frac{P(y) \cdot P(\frac{x_1}{y}) \cdot P(\frac{x_2}{y}) \cdot P(\frac{x_3}{y})}{P(x_1) \cdot P(x_2) \cdot P(x_3)}
$$

```bash
x1          x2          x3              y

-           -           -              Yes

-           -           -              No

-           -           -              Yes

-           -           -              No
```

Jadi :

- y = Yes. Misal Probabilitas yes ini = 0.60

$$
P(y=\text{Yes} \mid x_1, x_2, x_3) 
= \frac{ P(y=\text{Yes}) \cdot P(x_1 \mid y=\text{Yes}) \cdot P(x_2 \mid y=\text{Yes}) \cdot P(x_3 \mid y=\text{Yes}) }
       { P(x_1) \cdot P(x_2) \cdot P(x_3) }
$$

- y = No. Probabilitas no ini = 0.40

$$
P(y=\text{No} \mid x_1, x_2, x_3) 
= \frac{ P(y=\text{No}) \cdot P(x_1 \mid y=\text{No}) \cdot P(x_2 \mid y=\text{No}) \cdot P(x_3 \mid y=\text{No}) }
       { P(x_1) \cdot P(x_2) \cdot P(x_3) }
$$

Berarti hasil probabilitas adalahnya Yes.

## Naive Baye's Machine Learning Algorithm Implementation

https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html

````python
from sklearn.datasets import load_iris
````

````python
from sklearn.model_selection import train_test_split
````

````python
# disni dia mengembalikan nilai independent dan dependent
X,y = load_iris(return_X_y = True)
````

````python
X
````

````python
# 0 Setosa
# 1 Virgina
# 2 Versicolor
y
````

````python
# Data Splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# 70% training, 30% testing
# 80% training, 20% testing
````

````python
X_train
````

````python
from sklearn.naive_bayes import GaussianNB
````

````python
gnb=GaussianNB()
````

````python
gnb.fit(X_train,y_train)
````

````python
y_pred=gnb.predict(X_test)
````

````python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
````

````python
# setosa, versicolor, virginica
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# --- Contoh visualisasi Confusion Matrix ---
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, output_dict=True)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title(f"Confusion Matrix\nAccuracy = {acc:.2f}")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# --- Menampilkan Classification Report dalam bentuk DataFrame ---
import pandas as pd

report_df = pd.DataFrame(report).transpose()

plt.figure(figsize=(6,2))
sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Classification Report")
plt.show()
````
