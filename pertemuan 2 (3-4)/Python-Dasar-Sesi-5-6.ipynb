## 1. Pengenalan Syntax

**Syntax** adalah aturan penulisan program agar bisa dijalankan oleh Python.

Contoh dasar:

```python
print("Halo Dunia")
# Ini adalah komentar
```

Python menggunakan **indentasi (spasi)** untuk menandai blok kode, misalnya:

```python
umur = 20
if umur >= 18:
    print("Dewasa")
```


## 2. Tipe Data

Tipe data adalah jenis nilai yang disimpan dalam variabel.

Contoh tipe data umum:

```python
# Angka
x = 10          # int
y = 3.14        # float
z = 2 + 3j      # complex

# Teks
nama = "Vanya"  # str

# Boolean
lulus = True    # bool
```

Fungsi `type()` digunakan untuk cek tipe data:

```python
print(type(x))   # <class 'int'>
print(type(y))   # <class 'float'>
print(type(nama))# <class 'str'>
```


## 3. Variabel

Variabel digunakan untuk **menyimpan data**.

```python
nama = "Vanya"
umur = 22
ipk = 3.85

print(nama)
print(umur)
print(ipk)
```

Penamaan variabel:
- Harus diawali huruf/underscore
- Tidak boleh pakai spasi
- Case-sensitive (`nilai` â‰  `Nilai`)


## 4. Integer

Tipe data `int` menyimpan angka bulat (tanpa koma).

```python
x = 10
y = -5
z = 0

print(x + y)
print(type(z))  # <class 'int'>
```


## 5. Number (int, float, complex)

### Tipe Number:
- `int` âžœ angka bulat
- `float` âžœ angka desimal
- `complex` âžœ bilangan kompleks (a + bj)

```python
a = 7
b = 3.14
c = 2 + 4j

print(type(a))  # int
print(type(b))  # float
print(type(c))  # complex
```


## 6. List, Tuple dan Dictionary

### List: Kumpulan data yang bisa diubah (mutable)
```python
buah = ["apel", "jeruk", "mangga"]
buah.append("pisang")
print(buah[1])        # jeruk
print(len(buah))      # 4
```

### Tuple: Kumpulan data tetap (immutable)
```python
angka = (1, 2, 3)
print(angka[0])       # 1
```

### Dictionary: Pasangan key-value
```python
mahasiswa = {
    "nama": "Vanya",
    "umur": 22,
    "lulus": True
}
print(mahasiswa["nama"])
print(mahasiswa.get("umur"))
```
