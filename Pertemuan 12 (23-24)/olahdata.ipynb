## 1. Filtering dan Conditional Selection
Filtering digunakan untuk menyaring baris berdasarkan kondisi tertentu. 

Conditional Selection memungkinkan kita mengambil subset data berdasarkan kriteria logis (misalnya: usia > 30 atau jenis kelamin perempuan).

Seperti menyaring siswa dari kelas berdasarkan nilai di atas 80 dan jenis kelamin perempuan. Kita hanya ambil yang memenuhi syarat saja.

### Filter by 1 condition

```python
import pandas as pd

df = pd.read_csv('train.csv')
```

```python
# Filter 1 Kondisi
# menampilkan data yang umurnya atau Age > 30
df[df['Age'] > 30]
```

### Filter by 2 condition or more

```python
# Filter 2 kondisi atau lebih
# AND
# Menampilkan perempuan yang umur nya lebih dari 30
df[(df['Age'] > 20) & (df['Sex'] == 'female')]
```

```python
# Tampilkan penumpang wanita di kelas 1
df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
```

```python
# Tampilkan penumpang pria yang membayar tarif (Fare) di atas 100
df[(df['Sex'] == 'male') & (df['Fare'] > 100)]
```

```python
# AND
# Menampilkan umur yang dibawah 18 pada kelas 1
df[(df['Age'] < 18) & (df['Pclass'] == 1)]
```

```python
# OR
# Penumpang dengan umur di bawah 18 atau kelas 1
df[(df['Age'] < 18) | (df['Pclass'] == 1)]
```

Kenapa dia nampilin semuanya dan enggak sesuai dengan kondisi ?

Seharusnya dia nampilin Age < 18 dan Pclass == 1

Kita menggunakan OR maka hasilnya :

- Penumpang umur > 18 tapi kelas 1 → ditampilkan ✅

- Penumpang umur < 18 tapi kelasnya bukan 1 → tetap ditampilkan ✅
    
- Kalau dua-duanya terpenuhi → ya makin oke, tetap ditampilkan juga ✅

### .isin()
 digunakan untuk mengecek apakah nilai di dalam suatu kolom cocok dengan daftar nilai tertentu.

```python
# Struktur
DataFrame['kolom'].isin([nilai1, nilai2, ...])
```

- Akan menghasilkan True atau False untuk setiap baris.

- Digunakan untuk filtering (menyaring data) berdasarkan beberapa kondisi nilai.

```python
# Menyaring data dengan Pclass 1 atau 2
df_filtered = df[df['Pclass'].isin([1, 2])]

# Tampilkan hasil
df_filtered[['PassengerId', 'Pclass']].head()
```

### Kapan isin() digunakan ?

Saat ingin mencari baris dengan nilai tertentu di kolom tertentu.

Lebih ringkas daripada menulis banyak kondisi or:

```python
df[(df['Pclass'] == 1) | (df['Pclass'] == 2)]
```

```python
# Bisa diganti dengan
df[df['Pclass'].isin([1, 2])]
```

### Filter by String

```python
# Filter by string
df[df['Name'].str.contains("William")]
```

### Negasi (~)

Negasi artinya kebalikan dari kondisi yang diberikan.

Simbol negasi di Pandas adalah ~ (tilde).

Mirip kayak Boolean NOT.

Misalnya :
- df['Cabin'].isna() → menghasilkan baris yang Cabin-nya kosong (NaN)

- ~df['Cabin'].isna() → menghasilkan baris yang Cabin-nya TIDAK kosong

```python
# Baris yang Cabin-nya kosong
df[df['Cabin'].isna()]
```

```python
# Baris yang Cabin-nya TIDAK kosong (pakai negasi ~)
df[~df['Cabin'].isna()]
```

```python
# Semua penumpang yang BUKAN laki-laki
df[~(df['Sex'] == 'female')]  # Sama dengan df[df['Sex'] != 'male']
```

```python
# Penumpang yang bukan laki-laki
df[df['Sex'] != 'female']
```

### Multiple Condition

- AND: (cond1) & (cond2)

- OR: (cond1) | (cond2)

- Gunakan tanda kurung ( ) untuk tiap kondisi.

## 3. Menangani Missing Value
Missing value adalah data yang kosong atau tidak terisi, biasanya terlihat sebagai NaN (Not a Number) di pandas.

Missing value artinya ada data yang kosong atau hilang dalam sebuah kolom. Dalam Pandas, ini ditandai dengan NaN (Not a Number).

| Fungsi             | Kegunaan                                                      |
| ------------------ | ------------------------------------------------------------- |
| `dropna()`         | Menghapus baris/kolom yang mengandung `NaN`                   |
| `fillna()`         | Mengisi nilai yang hilang dengan nilai tertentu               |
| Imputasi sederhana | Mengisi `NaN` dengan rata-rata, median, modus, atau konstanta |


```python
import pandas as pd

# Load dataset
df = pd.read_csv('train.csv')
df.head()
```

```python
# Cek jumlah missing value di tiap kolom
df.isnull().sum()
```

### Dropna()

```python
# Hapus baris yang punya data kosong
# Untuk menghapus baris atau kolom yang memiliki NaN.
# Semua baris yang punya nilai kosong di kolom manapun akan dihapus.

df_bersih = df.dropna() #ngebersihin semuanya
```

```python
# Cek apakah masih ada NaN
df_bersih.isnull().sum()
```

```python
# Hapus baris jika kolom tertentu saja yang kosong
# Ini hanya menghapus baris kalau kolom Age-nya kosong.
# Artinya kamu menghapus baris hanya jika nilai pada kolom Age kosong (NaN), tanpa menyentuh kolom lain.
df_tanpa_age = df.dropna(subset=['Age'])
```

```python
print("Jumlah baris sebelum di-drop:", len(df))
print("Jumlah baris setelah drop Age NaN:", len(df_tanpa_age))
```

### Fillna()

```python
# Isi missing value kolom Age dengan nilai rata-rata
rata_rata_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(rata_rata_age)

print("Jumlah missing value di kolom Age setelah fillna:", df['Age'].isnull().sum())
print("Nilai rata-rata Age yang digunakan:", rata_rata_age)
```

```python
# Cek apakah NaN sudah berubah jadi mean
df[df['Age'] == rata_rata_age].head()
```

```python
#  Isi missing value kolom Embarked dengan 'S'
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df[df['Embarked'].notnull()]
```

| Fungsi     | Tujuan                                    | Risiko                                |
| ---------- | ----------------------------------------- | ------------------------------------- |
| `dropna()` | Menghapus data kosong                     | Bisa kehilangan banyak data penting   |
| `fillna()` | Mengisi data kosong dengan nilai tertentu | Bisa bias kalau nilai pengganti salah |



## 4. Menambah dan Mengubah Kolom

Kita bisa menambah kolom baru dari perhitungan, atau mengubah isi kolom yang sudah ada.

Analogi nya : Menambahkan kolom ‘Total Nilai’ dari jumlah nilai UTS dan UAS.

### Menambah Kolom Baru

```python
import pandas as pd

# Load dataset
df = pd.read_csv('train.csv')
df.head()
```

```python
# Kolom baru bernama FamilySize = SibSp + Parch + 1
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1 
df.head()
```

```python
# Kolom IsAlone → 1 jika sendirian, 0 jika ada keluarga
df['IsAlone'] = 0
df.loc[df['FamilySize'] == 1, 'IsAlone'] = 1
df.head()
```

### Mengubah Nilai di Kolom yang Sudah Ada

```python
# Contoh: Ubah isi kolom Sex dari 'male'/'female' → 0/1
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df.head()
```

```python
# atau bisa
df['Sex'] = df['Sex'].replace(['male', 'female'], [0, 1])
df.head()
```

```python
# Binning Kolom Age Menjadi Kategori
def age_category(age):
    if age < 12:
        return 'Child'
    elif age < 18:
        return 'Teen'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['AgeCategory'] = df['Age'].apply(age_category)
df
```

## 5. Rename dan Reorder Columns

Kadang kita perlu mengganti nama kolom agar lebih deskriptif atau menyusun ulang kolom agar lebih rapi.

Analogi: Mengganti "Nilai" jadi "Skor", atau menyusun kolom dari yang terpenting.

### Rename

Mengganti nama kolom agar lebih deskriptif, konsisten, atau mudah dipahami.

```python
# Bentuk strukturnya
DataFrame.rename(columns={'nama_lama': 'nama_baru'}, inplace=False)
```

- columns: dictionary yang berisi pasangan kolom_lama: kolom_baru.

- inplace=True: langsung mengubah di DataFrame asli.

- inplace=False: mengembalikan DataFrame baru (default).

```python
import pandas as pd

# Load dataset
df = pd.read_csv('train.csv')
df.head()
```

```python
# mengganti kolom 'Sex' menjadi 'Gender' dan 'Fare' menjadi 'TicketPrice'
df_renamed = df.rename(columns={'Sex': 'Gender', 'Fare': 'TicketPrice', 'PassengerId': 'ID'})

# hasil
df_renamed[['ID', 'Gender', 'TicketPrice']].head()
```

### Reordering Columns

Membuat tampilan DataFrame lebih logis atau fokus hanya pada kolom-kolom penting.

Buat ulang urutan kolom dengan menyusun ulang nama-nama kolom.

```python
# Tampilkan kolom yang tersedia terlebih dahulu
print(df.columns.tolist())
```

```python
# Setelah rename 'Sex' menjadi 'Gender' dan 'Fare' menjadi 'TicketPrice'
df_renamed = df.rename(columns={'Sex': 'Gender', 'Fare': 'TicketPrice'})
df_renamed.head()
```

```python
# Reorder kolom sesuai kebutuhan
df_reordered = df_renamed[['PassengerId', 'Name', 'Gender', 'Age', 'Pclass', 'TicketPrice']]

# Tampilkan hasil
df_reordered.head()
```
