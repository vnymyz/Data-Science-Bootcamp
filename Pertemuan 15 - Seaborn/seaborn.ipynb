### Seaborn
- Seaborn adalah library Python untuk visualisasi data yang dibangun di atas Matplotlib.

- Tujuannya: bikin grafik lebih cantik, rapi, dan informatif dengan sedikit kode.

- Punya style bawaan yang enak dilihat (warna, grid, font sudah otomatis rapi).

- Punya fungsi khusus untuk statistical plots seperti heatmap, boxplot, violinplot, dan pairplot.

| Perbedaan             | **Matplotlib**                                                                    | **Seaborn**                                                                           |
| --------------------- | --------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------- |
| **Tingkat Abstraksi** | Low-level (lebih detail, banyak pengaturan manual).                               | High-level (otomatis rapi, fokus ke insight data).                                    |
| **Fokus**             | Umum untuk semua jenis grafik.                                                    | Statistik & analisis data (terutama data tabular seperti Pandas DataFrame).           |
| **Gaya Bawaan**       | Standar dan sederhana (perlu di-style manual).                                    | Stylish & modern secara default.                                                      |
| **Dukungan Data**     | Bisa pakai array, list, DataFrame, tapi butuh coding lebih banyak untuk grouping. | Integrasi langsung dengan Pandas DataFrame dan otomatis grouping/agregasi.            |
| **Plot Statistik**    | Harus hitung sendiri statistiknya lalu gambar.                                    | Bisa langsung buat plot statistik dengan satu baris kode (`barplot`, `boxplot`, dll). |
| **Warna & Palet**     | Default 1 warna, harus set manual.                                                | Banyak palet warna bawaan (`sns.color_palette("coolwarm")`, dll).                     |


### Analogi

- Matplotlib itu seperti kanvas kosong: kamu harus gambar sendiri garis, warna, legenda, dsb.

- Seaborn itu seperti aplikasi desain siap pakai: tinggal pilih jenis grafik, data, dan warna, hasilnya langsung rapi.

### Scatter Plot
Scatterplot adalah grafik yang menampilkan titik-titik di koordinat x dan y, di mana:

- Sumbu x → Variabel pertama (independen)

- Sumbu y → Variabel kedua (dependen atau respon)

Tujuannya: melihat hubungan (relationship) antara dua variabel numerik.

### Kapan Scatterplot Digunakan?

1. Melihat korelasi atau pola

- Contoh: Apakah semakin besar total_bill, semakin besar tip?

- Bisa lihat linear, non-linear, atau tidak ada pola sama sekali.

2. Mendeteksi outlier

- Titik yang jauh dari cluster utama terlihat jelas.

3. Membandingkan kategori dengan warna/hue

- Bisa pakai hue, style, atau size untuk membedakan grup dalam dataset.

4. Eksplorasi data awal

- Cocok untuk EDA (Exploratory Data Analysis) sebelum modeling.

```python
# SCATTER PLOT
# create a scatter plot with tips dataset


sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

### Line Plot
Line plot adalah grafik yang menampilkan garis yang menghubungkan titik-titik data di sumbu x dan y.

- X-axis → biasanya variabel berurutan atau time/urutan

- Y-axis → variabel numerik yang ingin dilihat perubahannya

Tujuan utama: melihat tren atau perubahan nilai dari waktu ke waktu atau dari urutan tertentu.

### Analogi Sederhana

- Kamu mencatat suhu udara setiap jam di rumahmu.

- Kamu ingin tahu apakah siang lebih panas daripada pagi.

- Kamu gambar titik-titik suhu tiap jam → hubungkan dengan garis.

- Garis itu menunjukkan tren kenaikan atau penurunan suhu sepanjang hari.

### Kapan dipakai Line Plot ?

1. Time Series / Data Berurutan:

- Misal: suhu tiap jam, harga saham tiap hari, jumlah penumpang tiap bulan.

2. Melihat Tren:

- Apakah nilai naik, turun, atau stabil seiring waktu.

3. Membandingkan beberapa kategori seiring waktu:

- Bisa pakai hue untuk membandingkan grup (misal laki-laki vs perempuan, produk A vs B).

4. Eksplorasi Data:

- Lihat pola dan perubahan data numerik secara visual.

```python
# Line Plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line Plot of Total Bill vs Size")
plt.show()
```

### Bar Plot

- Barplot adalah grafik batang vertikal atau horizontal yang menunjukkan nilai numerik (y) untuk setiap kategori (x).

- X-axis → kategori (misal hari, jenis kelamin, produk)

- Y-axis → nilai numerik (misal rata-rata, total, count)

- Di Seaborn, sns.barplot() secara default menghitung rata-rata dari kolom numerik untuk tiap kategori, tapi bisa diganti dengan fungsi lain (estimator=np.sum, dll).

### Analogi 

- Kamu punya data jumlah kue terjual tiap hari.

- Kamu ingin tahu hari mana yang paling banyak kue terjual.

- Kamu buat batang untuk tiap hari, tingginya = rata-rata penjualan.

- Batang yang lebih tinggi → berarti penjualan lebih banyak.

### Kapan digunakan ?
1. Membandingkan kategori:

- Misal: rata-rata tip per hari (day vs tip)

- Misal: jumlah pelanggan per restoran

2. Mengevaluasi tren numerik per kategori:

- Lihat kategori mana yang lebih tinggi atau lebih rendah

3. EDA / Visualisasi Statistik Sederhana:

- Cocok untuk ringkasnya insight, misal perbandingan rata-rata, total, atau hitung (count)

```python
# Categorical Plots
# Bar plot
# example of categorical plot is sex, smoker, day, time
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Bar Plot of Total Bill by Day")
plt.show()
```

### Boxplot
Boxplot (atau kotak-garis) adalah grafik yang menunjukkan distribusi data numerik melalui lima ringkasan statistik:

- Minimum (kecuali outlier)

- Kuartil 1 (Q1 / 25%)

- Median (Q2 / 50%)

- Kuartil 3 (Q3 / 75%)

- Maximum (kecuali outlier)

- Outlier ditandai sebagai titik-titik terpisah di luar whisker.

- Box = Q1–Q3, garis di tengah = median, whisker = rentang data (tidak termasuk outlier).

Tujuannya:

- Melihat distribusi, spread, median, dan outlier dalam satu plot.

### Analogi

- Kamu punya nilai ujian teman-teman.

- Kamu ingin tahu: median nilai, range, dan siapa yang sangat jomplang (outlier).

- Boxplot → kotak untuk nilai tengah, garis whisker untuk range normal, titik di luar kotak → teman dengan nilai ekstrem.

### Kapan Menggunakan Boxplot

1. Mengecek distribusi data numerik per kategori

- Misal: total_bill per hari, tip per waktu makan

2. Mendeteksi outlier

- Titik yang jauh dari kotak → data yang tidak biasa

3. Membandingkan distribusi antar kategori

- Bisa pakai hue untuk membedakan grup tambahan

```python
# Box Plot
# good for seeing outliers
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

### Violin Plot

Violin plot adalah kombinasi boxplot + distribusi data (density plot).

- Menunjukkan distribusi data dan nilai statistik (median, quartile) secara visual.

- Bentuknya seperti biola (symmetrical), makin lebar → makin banyak data di nilai tersebut.

- Median biasanya ditandai garis di tengah, kadang juga menampilkan quartile.

Tujuan utama:

- Melihat distribusi data lebih detail daripada boxplot

- Mudah melihat apakah data skewed, bimodal, atau normal

### Analogi

- Kamu punya tip yang diberikan pelanggan.

- Ingin tahu bukan hanya median, tapi juga seberapa banyak orang memberi tip di range tertentu.

- Violin plot → bentuk seperti “biola”:

    - Bagian lebar → banyak orang memberi tip di angka itu

    - Bagian sempit → sedikit orang memberi tip di angka itu

- Garis tengah → median

### Kapan digunakan ?

1. Melihat distribusi data numerik per kategori

- Lebih informatif dibanding boxplot jika distribusi kompleks

2. Mendeteksi pola multi-modal

- Kalau ada dua puncak distribusi, violin plot terlihat jelas

3. Membandingkan kategori

- Bisa ditambah hue untuk melihat distribusi berbeda antar grup

### Jawaban Solusi

```python
# 1. Scatterplot: total_bill vs tip dengan hue = sex
plt.figure(figsize=(6,4))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Scatterplot Total Bill vs Tip berdasarkan Jenis Kelamin")
plt.show()
```

```python
# 2. Barplot: rata-rata tip per day
# ini dia bakal otomatis ngitung rata2
# Agregasi rata-ratanya dihitung otomatis oleh Seaborn, karena sns.barplot() default-nya pakai fungsi estimator=np.mean.
plt.figure(figsize=(6,4))
sns.barplot(data=tips, x="day", y="tip")
plt.title("Rata-rata Tip per Hari")
plt.show()
```

```python
# 3. Countplot: jumlah transaksi berdasarkan size
plt.figure(figsize=(6,4))
sns.countplot(data=tips, x="size")
plt.title("Jumlah Transaksi berdasarkan Jumlah Orang di Meja")
plt.show()
```

```python
# 4. Boxplot: total_bill per day dengan hue = sex
plt.figure(figsize=(6,4))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Distribusi Total Bill per Hari berdasarkan Jenis Kelamin")
plt.show()
```

```python
# 5. Violinplot: distribusi tip berdasarkan time
plt.figure(figsize=(6,4))
sns.violinplot(data=tips, x="time", y="tip")
plt.title("Distribusi Tip berdasarkan Waktu Makan")
plt.show()
```

```python
# 6. Relplot: scatter total_bill vs tip dengan col = smoker
sns.relplot(data=tips, x="total_bill", y="tip", col="smoker", kind="scatter", height=4, aspect=1)
plt.suptitle("Scatterplot Total Bill vs Tip (Dipisah berdasarkan Smoker)", y=1.05)
plt.show()
```

```python
# 7. Heatmap: korelasi kolom numerik
plt.figure(figsize=(5,4))
sns.heatmap(tips[["total_bill", "tip", "size"]].corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap Korelasi")
plt.show()
```

```python
# 8. Pairplot: kolom numerik dengan hue = smoker
sns.pairplot(tips[["total_bill", "tip", "size", "smoker"]], hue="smoker")
plt.suptitle("Pairplot Kolom Numerik dengan Warna berdasarkan Smoker", y=1.02)
plt.show()
```

```python
# 9. Histogram: distribusi total_bill dengan KDE
plt.figure(figsize=(6,4))
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("Distribusi Total Bill dengan KDE")
plt.show()
```

```python
# 10. Swarmplot: persebaran tip per day dengan hue = time
plt.figure(figsize=(6,4))
sns.swarmplot(data=tips, x="day", y="tip", hue="time")
plt.title("Persebaran Tip per Hari berdasarkan Waktu Makan")
plt.show()
```
