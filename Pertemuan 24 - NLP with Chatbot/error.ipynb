```python
import pandas as pd
import numpy as np

# Pretend dataset
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Evan"],
    "Age": [25, 30, np.nan, 22, 29],
    "Score": [88, 92, 85, 90, None]
}

df = pd.DataFrame(data)
print("=== Original Data ===")
print(df)

# Try to calculate average age
avg_age = df["Ages"].mean()   # 🐞 Oops! Column name is wrong
print("\nAverage Age:", avg_age)

# Fill missing scores
df["Score"] = df["Score"].fillna(df["Score"].mean)

# Compute custom metric
df["Performance"] = df["Score"] / df["Age"] * 10

# Sort by Performance
df.sort_values("Performance", ascending=False, inplace=True)

print("\n=== Processed Data ===")
print(df)

# Export to CSV
df.to_csv("processed_data.csv")
print("\nData saved successfully!")
```
